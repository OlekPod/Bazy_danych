wykres podanej funkcji dla argumentów z podanych przedziałów.

```python
def fun(x):
    return(x*x + 5)
```

```python
#zd 3
pairs = [(-1,1), (-6,6), (0,5)]

for pair in pairs:
    y = []
    xx = np.linspace(pair[0], pair[1], 1000)
    for x in xx:
        #y.append(x*x + 5)
        y.append(fun(x))
    
    plt.figure(id(pair))
    plt.title("Przedział " + str(pair))
    plt.xlabel("x")
    plt.ylabel("x^2 + 5")
    plt.plot(xx,y)
```
